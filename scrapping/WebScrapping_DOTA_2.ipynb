# DOTA 2 Heroes Attributes

![dota2](https://cdns.klimg.com/bola.net/library/upload/21/2018/09/hero-dota-2_61002e6.jpg)

Goals of this project is to scrape the atribut of DOTA2 Heroes dataset into CSV file from [DOTA 2 Fandom Website](https://dota2.fandom.com/wiki/Table_of_hero_attributes). 
The Scrapping using BeautifulSoup from python library.


##**WEBSCRAPPING**

```python
#import libraries

import pandas as pd
import numpy as np
from urllib.request import urlopen
from bs4 import BeautifulSoup
import warnings
import os
warnings.filterwarnings('ignore')
```

```python
#write the url into variable
url = 'https://dota2.fandom.com/wiki/Table_of_hero_attributes'
#open the url 
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
#Get the table of contents
rows = soup.find_all('tr')

all_rows = []
for row in rows:
  row_td = row.find_all('td')
  row_td = str(row_td)
  row_td = BeautifulSoup(row_td, 'lxml').get_text()
  row_td = row_td.replace('\r', '')
  row_td = row_td.replace('\n', '')
  row_td = row_td.replace(' ','')

  all_rows.append(row_td)

print(all_rows)
```

the content of table are written in numpy array

```python
#convert the numpy arrray to dataframe
contents = pd.DataFrame(all_rows)
contents.tail(5)
```

```python
contents = contents[0].str.split(',', expand=True)
contents
```

```python
contents[0] = contents[0].str.strip('[]')
contents
```

```python
contents[28] = contents[28].str.strip('[]')
contents
```

```python
#Because there is unused rows in rows 123 to 138
#Drop the unused rows
contents = contents.drop(contents.index[123:])
contents
```

```python
#drop the first cow with None value
contents = contents.drop(contents.index[0])
contents
```

```python
#Scrapping the header for heroes attributes
col_labels = soup.find_all('th')
hero_header = []
col_name = str(col_labels)
col_name = col_name.replace('\r', '')
col_name = col_name.replace(' ', '')
col_name = col_name.replace('\n', '')
col_name = col_name.replace('[', '')
col_name = col_name.replace(']', '')
cleantext2 = BeautifulSoup(col_name, 'lxml').get_text()
hero_header.append(cleantext2)
print(hero_header)
```

```python
#convert numpy array into dataframe
headers = pd.DataFrame(hero_header)
headers
```

```python
#split the string with ','
headers = headers[0].str.split(',', expand=True)
headers
```

```python
#merge the content and headers
frames = [headers, contents]
final = pd.concat(frames)
final
```

```python
#make the first row to column name
final = final.rename(columns=final.iloc[0])
final
```

```python
#drop unused rows
final = final.drop(final.index[0])

#drop unused columns
final.drop(final.columns[29], axis = 1, inplace=True)
final.drop('A', axis=1, inplace=True)

#check the final data
final
```

```python
#Scrapping the for additional columns 'MAINATT' (MAINATT is on images)
import urllib.request

thepage = urllib.request.urlopen(url)
sup = BeautifulSoup(thepage)

raw_att = [i['title'] for i in soup.find_all('a', title=True)]
print(raw_att)
```

```python
#convert the numpy arrray to dataframe
raw_att = pd.DataFrame(raw_att)
raw_att
```

```python
#Because there is unused rows in rows 377 to end
#Drop the unused rows
raw_att = raw_att.drop(raw_att.index[377:])
raw_att
```

```python
#Because there is unused rows in rows first to 10
#Drop the unused rows
raw_att = raw_att.drop(raw_att.index[:11])
raw_att
```

```python
#define the reference list to filter raw_att
list_att = ['Strength', 'Agility', 'Intelligence']
list_att = pd.DataFrame(list_att)
list_att
```

```python
#check if the list in list_att, the value will be assign in the list prim_att
prim_att = raw_att[raw_att[0].isin(list_att[0])]

#reindexing from 1
prim_att.index = np.arange(1, len(prim_att)+1)

prim_att
```

```python
#add column 'Attribute'
final['Attribute'] = prim_att
final
```

```python
#define the clean dataframe
df = final.copy()
df
```

```python
#export the file
final.to_csv('dota2.csv', index = False)
```
